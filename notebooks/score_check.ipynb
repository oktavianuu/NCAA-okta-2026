```python
data = pd.read_csv("ncaa_2026_data.csv")
```

```python
def find_matchup(team1, team2, df):
    """
    Find the row for the matchup between team1 and team2.
    Returns the row if found, else None.
    """
    # Try original order
    mask = (df['Team_A'] == team1) & (df['Team_B'] == team2)
    # If not found, try reversed order
    if not mask.any():
        mask = (df['Team_A'] == team2) & (df['Team_B'] == team1)
    
    result = df[mask]
    return result if not result.empty else None

# Example usage
matchup = find_matchup("Duke", "Siena", data)
print(matchup)
```
